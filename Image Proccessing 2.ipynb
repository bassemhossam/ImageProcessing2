```python
img = cv2.imread('image2_folder/Square-circle.png',0)
```

1.2 Square Kernel

```python
sq_kernel=cv2.getStructuringElement(cv2.MORPH_RECT, (10, 10))
square_dilate_img=cv2.dilate(img, sq_kernel, iterations=1)
cv2.imwrite('square-circle-1.png',square_dilate_img)
```

1.3 Circle Kernel

```python
circ_kernel=cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (10, 10))
circle_dilate_img=cv2.dilate(img, circ_kernel, iterations=1)
cv2.imwrite('square-circle-2.png',circle_dilate_img)
```

1.5 Both the kernels has dilated the image where it increased the white region in the image but the only observable difference is the shape of edges between white and black where they appear more sharp and linear with the square kernel(the square's shape is preserved but not the black circle) but the edges appear more curved when the circular kernel is used(edges of the square are curved now  but maintaining the shape of the black circle) 
Also increasing the kernel size increases the dilation in the image

Opening and Closing

7.a.1 Extracting circles

```python
img7 = cv2.imread('image2_folder/Circle_and_Lines.png',0)
kernel_open=cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(12,12))
circles =cv2.morphologyEx(img7,cv2.MORPH_OPEN,kernel_open)
cv2.imwrite('circles.png',circles)
plt.imshow(circles,cmap='gray')
plt.show()
```

7.a.2 Extracting lines

```python
lines= cv2.subtract(img7,circles)
# new_kernel=cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
# lines =cv2.morphologyEx(lines,cv2.MORPH_OPEN,new_kernel)
new_kernel=cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
lines =cv2.morphologyEx(lines,cv2.MORPH_OPEN,new_kernel)
cv2.imwrite('lines.png',lines)

plt.imshow(lines,'gray')
plt.show()
```

7.a To seperate the circles a circular kernel was used because the opening technique tends to retain forground shapes that are similar to the structuring element as opposed to using a rectangular kernel which couldn't really capture the circles.
To get the lines the circles were subtracted from the original image and then opening was used to make the lines clearer.
This method is limited as the shape of the seperated objects is similar to the shape of the structuring element so it can't clearly seperate objects of different sizes or different orientatins .
As the size of the circular kernel increased it started to delete the lines from the image but with smaller kernel sizes(below 8x8) the lines weren't completely removed.

7.b.1 Counting Circles

```python
# Set up the detector with default parameters.
output = cv2.connectedComponentsWithStats(circles)
count=output[0]-1
print(count)
circles_labeled=circles.copy()
centroids=output[3]
# circles_labeled=cv2.cvtColor(circles_labeled,cv2.COLOR_GRAY2BGR)
for i in range(1,count+1):
    cX=int(centroids[i][0])
    cY=int(centroids[i][1])
    cv2.putText(circles_labeled,str(i),(cX-5,cY+5),cv2.FONT_HERSHEY_SIMPLEX, 0.3, 120)
    
plt.imshow(circles_labeled,cmap='gray')
plt.show()
```

8.a Computing skeleton

```python
def get_skeleton(in_image):
    in_image=cv2.bitwise_not(in_image)
    out_image = np.zeros(in_image.shape,dtype=np.uint8)
    kernel = cv2.getStructuringElement(cv2.MORPH_CROSS, (5,5))
    finished = False
    while not finished:
        eroded = cv2.erode(in_image,kernel)
        temp = cv2.dilate(eroded,kernel)
        temp = cv2.subtract(in_image,temp)
       
        out_image = cv2.bitwise_or(out_image,temp)
        
        in_image = eroded.copy()
     
        if cv2.countNonZero(in_image) == 0:
            finished = True
   
    return out_image





girl = cv2.imread('image2_folder/Girl.png',0)
girl_skeleton=get_skeleton(girl)
Dog = cv2.imread('image2_folder/Dog.png',0)
Dog_skeleton=get_skeleton(Dog)
hot_ballon = cv2.imread('image2_folder/HotBallon.png',0)
hot_ballon_skeleton=get_skeleton(hot_ballon)
res_plot5=plt.figure(figsize=(15,15))
left5=res_plot5.add_subplot(1,3,1)
left5.set_title('Girl')
left5.imshow(girl_skeleton,cmap='gray')
mid5=res_plot5.add_subplot(1,3,2)
mid5.set_title('Dog')
mid5.imshow(Dog_skeleton,cmap='gray')
right5=res_plot5.add_subplot(1,3,3)
right5.set_title('Hot Ballon')
right5.imshow(hot_ballon_skeleton,cmap='gray')
plt.show()
```

8.a This method continously keeps thinning and eroding the image from the boundary until no more thinning is possible and approximately obtain the skeleton,it uses a cross kernel with size 5x5 and when the size increases tthe skeleten becomes thicker and  the lines become more non uniform. 

8.b&c Get center of mass and area

```python
def get_center_and_area(input_image):
    input_image = cv2.bitwise_not(input_image)
    ret,input_image=cv2.threshold(input_image,50,255,cv2.THRESH_BINARY)
    kernel_close=cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))

    input_image=cv2.morphologyEx(input_image,cv2.MORPH_CLOSE,kernel_close)

    out = cv2.connectedComponentsWithStats(input_image)
    centroids = out[3]
    stats = out[2]
    labeled_shapes = out[1]
    return labeled_shapes,centroids[1], stats[1, 4]


girl_lables,girl_centroid,girl_area=get_center_and_area(girl)
print('Girl center of mass:',girl_centroid)
print('Girl Area:',girl_area)


dog_lables,dog_centroid,dog_area=get_center_and_area(Dog)
print('Dog center of mass:',dog_centroid)
print('Dog Area:',dog_area)


ballon_lables,ballon_centroid,ballon_area=get_center_and_area(hot_ballon)
print('Hot Ballon center of mass:',ballon_centroid)
print('Hot Ballon Area:',ballon_area)


```

8.b&c The image is first inverted ,thresholded and then a morphological close is applied on it to get a clear binary image,then it is passed to cv2.
